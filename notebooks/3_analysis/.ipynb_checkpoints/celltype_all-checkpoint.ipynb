```r
markerHSC <- getMarkerFeatures(
  ArchRProj = proj,
  useMatrix = "PeakMatrix",
  groupBy = "tissue_celltype",
  testMethod = "wilcoxon",
  bias = c("TSSEnrichment", "log10(nFrags)"),
  useGroups = "BoneMarrow_HSC_MPP",
  bgdGroups = "MobilizedBlood_HSC_MPP"
)

# Plot results
pma <- plotMarkers(
  seMarker = markerHSC, 
  name = "BoneMarrow_HSC_MPP", 
  cutOff = "FDR <= 0.05 & abs(Log2FC) >= 1",
  plotAs = "MA"
)

pv <- plotMarkers(
  seMarker = markerHSC, 
  name = "BoneMarrow_HSC_MPP", 
  cutOff = "FDR <= 0.05 & abs(Log2FC) >= 1",
  plotAs = "Volcano"
)
```

```r
plotPDF(pma, pv, name = "HSC-Differential-Peaks", width = 10, height = 5)

# Get significant peaks
hscPeaks <- getMarkers(markerHSC, cutOff = "FDR <= 0.05 & abs(Log2FC) >= 1")
```

```r
# Add motif annotations
proj <- addMotifAnnotations(
  ArchRProj = proj,
  motifSet = "cisbp",
  name = "Motif",
  force = TRUE
)

# Motif enrichment in differential peaks
enrichMotifs <- peakAnnoEnrichment(
  seMarker = markerHSC,
  ArchRProj = proj,
  peakAnnotation = "Motif",
  cutOff = "FDR <= 0.05 & abs(Log2FC) >= 1"
)

# Plot
heatmapEM <- plotEnrichHeatmap(
  enrichMotifs, 
  n = 25, 
  transpose = TRUE
)
```

```r
library(ggplot2)
library(scales)

# Extract relevant data
enrichDF <- data.frame(
  name = rownames(enrichMotifs),
  mlog10Padj = assays(enrichMotifs)$mlog10Padj[,1],
  enrichment = assays(enrichMotifs)$Enrichment[,1],
  compareFreq = assays(enrichMotifs)$CompareFrequency[,1],
  backgroundFreq = assays(enrichMotifs)$BackgroundFrequency[,1]
)

# Filter and sort
enrichDF <- enrichDF[!is.na(enrichDF$mlog10Padj), ]
topMotifs <- head(enrichDF[order(enrichDF$mlog10Padj, decreasing = TRUE), ], 25)

# Enrichment as color
topMotifs$name <- factor(topMotifs$name, 
                         levels = topMotifs$name[order(topMotifs$mlog10Padj)])

p <- ggplot(topMotifs, aes(x = name, y = mlog10Padj, fill = enrichment)) +
  geom_bar(stat = "identity") +
  scale_fill_gradient2(low = "blue", mid = "white", high = "darkred", 
                       midpoint = 1, name = "Fold\nEnrichment") +
  coord_flip() +
  theme_minimal() +
  theme(axis.text.y = element_text(size = 7)) +
  labs(x = "", y = "-log10(Adjusted P-value)",
       title = "Top 25 Enriched Motifs in HSC",
       subtitle = "Color indicates fold enrichment")

ggsave("Motif-Enrichment-HSC-enhanced.pdf", p, width = 10, height = 10)
print(p)
```

```r
#### 1) Marker peaks – collab vs granja in HSCs ####
markerPeaks <- getMarkerFeatures(
  ArchRProj  = proj_hsc,
  useMatrix  = "PeakMatrix",
  groupBy    = "Tissue",
  testMethod = "wilcoxon",
  bias       = c("TSSEnrichment","log10(nFrags)"),
  useGroups  = "BoneMarrow",
  bgdGroups  = "MobilizedBlood"
)

# Extract GRanges of significant peaks
bmPeaks <- getMarkers(
  markerPeaks,
  cutOff   = "FDR <= 0.05 & Log2FC >= 1",
  returnGR = TRUE
)$BoneMarrow

mbPeaks <- getMarkers(
  markerPeaks,
  cutOff   = "FDR <= 0.05 & Log2FC <= -1",
  returnGR = TRUE
)$MobilizedBlood
```

```r
# Add motif annotations (cisBP)
proj_hsc <- addMotifAnnotations(
  ArchRProj = proj_hsc,
  motifSet   = "cisbp",
  name       = "Motifs",
    force=TRUE
)

# rebuild the PeakMatrix so it matches your new PeakSet
proj_hsc <- addPeakMatrix(proj_hsc, force = TRUE)

# Wnrichment of motifs in marker peaks
proj_hsc <- addBgdPeaks(proj_hsc, force=TRUE)
```

```r
# Enrichment of motifs in marker peaks (corrected)

# enrichment test
motifEnr <- peakAnnoEnrichment(
  seMarker    = markerPeaks,
  ArchRProj   = proj_hsc,
  peakAnnotation = "Motif"
)

# top motifs for each group
head(motifEnr$enrichment$BoneMarrow, 10)
head(motifEnr$enrichment$MobilizedBlood, 10)

# heat-map of all enriched motifs
plotEnrichHeatmap(
  enrichment = motifEnr,
  n          = 20,
  transpose  = TRUE
)

# chromVAR deviations + differential testing
proj_hsc <- addDeviationsMatrix(
  ArchRProj      = proj_hsc,
  peakAnnotation = "Motifs",
  force          = TRUE
)

# differential deviations
diffDev <- differentialDeviations(
  ArchRProj  = proj_hsc,
  groupBy    = "Tissue",
  testMethod = "wilcoxon",
  useGroups  = "BoneMarrow",
  bgdGroups  = "MobilizedBlood"
)

# Top differential motifs by absolute z‐score
head(diffDev[ order(abs(diffDev$z), decreasing=TRUE), ], 20)
```

```r
library(ComplexHeatmap)

# top 20 motifs by enrichment
topMotifs <- rownames(motifEnr$BoneMarrow)[1:20]

# get chromVAR deviation z‐scores per group
devMat <- getGroupMatrix(
  ArchRProj = proj_hsc,
  groupBy   = "Tissue",
  useMatrix = "MotifMatrix"
)

# subset & scale
hmDat <- scale(t(devMat[topMotifs, ]))

Heatmap(
  hmDat,
  name      = "z-score",
  cluster_rows = TRUE,
  cluster_columns = TRUE,
  column_title = "Tissue",
  row_title    = "Top Motifs"
)
```
